## Dataset "healthcare-dataset-stroke-data.csv"

```python
#Leemos dataset descargado desde aquí: https://gist.github.com/aishwarya8615/d2107f828d3f904839cbcb7eaa85bd04#file-healthcare-dataset-stroke-data-csv
df_healthcare_stroke = pd.read_csv('datasets/healthcare-dataset-stroke-data.csv')
df_healthcare_stroke.head()
```

Podemos observar datos básicos a nivel salud y sociales de cada persona que se entienden en su mayoría. BMI se refiere al índice al "Body mass index", el índice de masa corporal.
